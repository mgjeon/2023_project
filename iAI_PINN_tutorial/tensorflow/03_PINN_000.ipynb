```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Plate_data = np.load('../Plate_data.npy')
Plate_data.shape
```

```python
loc = Plate_data[:, 0:2]
u = Plate_data[:, 2:3]
v = Plate_data[:, 3:4]
stress = Plate_data[:, 4:6]

loc.shape, u.shape, v.shape, stress.shape
```

```python
a = 10
b = 10
```

```python
plt.scatter(loc[:, 0], loc[:, 1], s=0.5, c='b')
plt.title('Points', fontsize=15)
plt.xlabel('x (mm)', fontsize = 14)
plt.ylabel('y (mm)', fontsize = 14)
plt.axis('square')
plt.xlim([0, a])
plt.ylim([0, b])
plt.show()
```

```python
color_legend = [[0, 0.182], [-0.06, 0.011], [-0.0022, 1.0], [-0.15, 0.45]]
title = ['x-displacement ($u$)', 'y-displacement ($v$)', '$\sigma_{xx}$', '$\sigma_{yy}$']

true_disp = np.hstack([u, v])
true_result = np.hstack([true_disp, stress])

plt.figure(figsize=(10, 8))
for idx in range(4):
    plt.subplot(2, 2, idx+1)
    plt.scatter(loc[:, 0], loc[:, 1], c=true_result[:, idx], cmap='rainbow', s=5)
    plt.clim(color_legend[idx])
    plt.title(title[idx])
    plt.xlabel('x (mm)', fontsize = 14)
    plt.ylabel('y (mm)', fontsize = 14)
    plt.axis('square')
    plt.xlim([0, a])
    plt.ylim([0, b])
    plt.colorbar()
plt.tight_layout()
plt.show()
```
