```python
a = [[1,2], [3,4]]
b = a
a[0] = [100, 1000]
a, b
```

```python
a = [[1,2], [3,4]]
b = a[:]
a[0] = [100, 1000]
a, b
```

```python
id(a), id(b)
```

```python
a = [[1,2], [3,4]]
b = a.copy()
a[0] = [100, 1000]
a, b
```

```python
id(a), id(b)
```

```python
a = [[1,2], [3,4]]
b = a.copy()
a[0][0] = 100
a, b
```

```python
id(a), id(b)
```

```python
id(a[0]), id(b[0])
```

```python
from copy import deepcopy
```

```python
a = [[1,2], [3,4]]
b = deepcopy(a)
a[0][0] = 100
a, b
```

```python
id(a), id(b)
```

```python
id(a[0]), id(b[0])
```
