```python
num = 8
```

```python
x1 = np.linspace(b.x1[0], b.x1[-1], num)
y1 = np.linspace(b.x1[0], b.x1[-1], num)
x1s, y1s = np.meshgrid(x1, y1)
z1s = x1s*0 + b.z1[0]
```

```python
x1s.shape, y1s.shape, z1s.shape
```

```python
x1s
```

```python
x0 = np.stack((x1s.flatten(), y1s.flatten(), z1s.flatten()), axis=1)
```

```python
x0.shape
```

```python
x0
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib ipympl
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121, projection='3d')
ax.set_xlim(b.x1[0], b.x1[-1])
ax.set_ylim(b.y1[0], b.y1[-1])
ax.set_zlim(b.z1[0], b.z1[-1])
for xb in x0:
    ax.scatter(xb[0], xb[1], xb[2])

for xl in xls:
    ax.plot(xl[:, 0], xl[:, 1], xl[:, 2])
    
plt.show()
```
