```python
import os 
os.environ['CUDA_VISIBLE_DEVICES'] = '0'

import torch
import numpy as np
from tools.model import Unet
```

```python
model = Unet().to('cuda')
```

```python
checkpoint = torch.load('/userhome/jeon_mg/workspace/project/CNN_NF2/NF2_notebooks/cnn/model_unet_b3_clr/model_10000.pt')
model.load_state_dict(checkpoint['model_state_dict'])
model.eval();
```

```python
inputs = np.load('data/inputs_11158.npy')
outputs = np.load('data/outputs_11158.npy')
```

```python
inputs = torch.Tensor(inputs).to('cuda')
labels = outputs

inputs.shape, labels.shape
```

```python
model_input = torch.permute(inputs, (0, 4, 3, 2, 1))
model_input.shape
```

```python
b_norm = 2500
```

```python
N = model_input.shape[0]
N
```

```python
bs = []
for i in range(N):
    model_input[i].shape
    b = model(model_input[i][None, ...])
    b = b.cpu().detach().numpy().transpose(0, 3, 2, 1, 4)*b_norm
    bs.append(b)
bs = np.array(bs)
bs.shape
```

```python
bs = bs[:, 0, ...]
bs.shape
```

```python
b2 = (bs**2).sum(-1).sum((1, 2, 3))
b2.shape
```

```python
b2[0]
```

```python
labels.shape
```

```python
B2 = ((labels*b_norm)**2).sum(1).sum((1, 2, 3))
B2.shape
```

```python
B2[0]
```

```python
eps = b2/B2 
eps[0]
```

```python
outputs_pot = np.load('data/outputs_pot_11158.npy')
```

```python
outputs_pot.shape
```

```python
Bpot2 = ((outputs_pot*b_norm)**2).sum(1).sum((1, 2, 3))
Bpot2.shape
```

```python
Bpot2[0]
```

```python
plt.plot(b2-Bpot2, label='prediction free energy')
plt.plot(B2-Bpot2, label='isee free energy')
plt.legend()
```
